# MiniSpecialFunctions Demo Notebook\n\nDemonstrates numerical computation of special functions\nusing the MiniSpecialFunctions Lean 4 package.

```python
# Gamma function special values\nimport math\nprint(f'Gamma(1)   = 1          (exact: {math.gamma(1)})')\nprint(f'Gamma(5)   = 24         (computed: {math.gamma(5)})')\nprint(f'Gamma(0.5) = sqrt(pi)   (computed: {math.gamma(0.5)})')\nprint(f'sqrt(pi)                 = {math.sqrt(math.pi)}')
```

```python
# Riemann zeta special values\nimport mpmath as mp\nprint(f'zeta(2)  = pi^2/6 = {mp.zeta(2)}')\nprint(f'pi^2/6              = {math.pi**2/6}')\nprint(f'zeta(4)  = pi^4/90 = {mp.zeta(4)}')\nprint(f'zeta(-1) = -1/12   = {mp.zeta(-1)}')
```
