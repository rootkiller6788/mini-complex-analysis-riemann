# Entire Function Exploration

Exploring entire functions: order of growth, Weierstrass products, Hadamard factorization.

```python
import numpy as np
import matplotlib.pyplot as plt

# Primary factor E_k(z)
def primary_factor(z, k):
    if k == 0:
        return 1 - z
    partial_sum = sum(z**m / m for m in range(1, k+1))
    return (1 - z) * np.exp(partial_sum)

# Weierstrass canonical product
def weierstrass_product(zeros, genus, z, N):
    result = 1.0
    for i, a in enumerate(zeros[:N]):
        if abs(a) > 1e-10:
            result *= primary_factor(z/a, genus)
        else:
            result *= z
    return result
```

```python
# sin(pi*z) = pi*z * prod_{n!=0} (1 - z/n)
def sin_pi_zeros(N):
    zeros = []
    for n in range(1, N+1):
        zeros.append(n)
        zeros.append(-n)
    return zeros

z_test = 0.5
product_val = weierstrass_product(sin_pi_zeros(50), 0, z_test, 100)
exact_val = np.sin(np.pi * z_test) / (np.pi * z_test)
print(f"z={z_test}: product={product_val:.6f}, exact={exact_val:.6f}")
```

```python
# Order of growth computation
def max_modulus(f, r, N=100):
    max_val = 0.0
    for k in range(N):
        theta = 2*np.pi*k/N
        z = r * np.exp(1j*theta)
        val = np.abs(f(z))
        max_val = max(max_val, val)
    return max_val

def estimate_order(f, r_values):
    log_log_M = []
    log_r = []
    for r in r_values:
        M = max_modulus(f, r)
        if M > 1:
            log_log_M.append(np.log(np.log(M)))
            log_r.append(np.log(r))
    if len(log_log_M) > 1:
        return (log_log_M[-1] - log_log_M[0]) / (log_r[-1] - log_r[0])
    return 0.0

f = np.exp
r_vals = [1.0, 2.0, 5.0, 10.0]
est_order = estimate_order(f, r_vals)
print(f"Estimated order of exp(z): {est_order:.4f} (exact: 1)")
```
