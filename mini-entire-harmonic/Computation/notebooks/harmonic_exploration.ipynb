# Harmonic Function Exploration

Exploring harmonic functions: Laplace equation, Poisson kernel, Dirichlet problem.

```python
import numpy as np
import matplotlib.pyplot as plt

# Poisson kernel P(r, theta)
def poisson_kernel(r, theta):
    return (1 - r**2) / (1 - 2*r*np.cos(theta) + r**2)

# Harmonic function from boundary data via Poisson integral
def poisson_integral(boundary, r, theta, N=1000):
    dt = 2*np.pi/N
    result = 0.0
    for k in range(N):
        t = k*dt
        result += poisson_kernel(r, theta - t) * boundary(t) * dt
    return result / (2*np.pi)
```

```python
# Check that Poisson of constant 1 yields 1
boundary_const = lambda t: 1.0
for r in [0.0, 0.25, 0.5, 0.75, 0.9]:
    val = poisson_integral(boundary_const, r, 0.0, N=500)
    print(f"r={r:.2f}: u(r,0) = {val:.6f}")
```

```python
# Solve Dirichlet problem for boundary f(theta) = cos(theta)
# Exact solution: u(r,theta) = r*cos(theta)
boundary_cos = lambda t: np.cos(t)
for r in [0.0, 0.3, 0.6, 0.9]:
    for theta in [0, np.pi/4, np.pi/2]:
        numeric = poisson_integral(boundary_cos, r, theta, N=500)
        exact = r*np.cos(theta)
        print(f"r={r:.2f}, theta={theta:.2f}: numeric={numeric:.4f}, exact={exact:.4f}")
```
